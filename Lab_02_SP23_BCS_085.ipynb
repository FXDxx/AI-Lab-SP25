```python
# A palindrome is a string which is same read forward or backwards.
def palindrome(word):
  temp = word[::-1]
  if temp.capitalize() == word.capitalize():
    return True
  else:
    return False

print(palindrome("alflals"))
```

```python
'''Write a python function that takes a list of N tuples as input and returns the perimeter of the
polygon. Remember that your code should work for any value of N.'''

def perimeter(list):
  length = len(list)
  perimeter = 0
  for i in range(0, length-1):
    dist = (((list[i][0]-list[i+1][0])**2)+ ((list[i][1]-list[i+1][1])**2))**0.5

    parimeter = perimeter+dist

  perimeter = perimeter + (((list[0][0]-list[length-1][0])**2)+((list[0][1]-list[length-1][1])**2))**0.5
  return parimeter

List = [(1,3),(2,7),(3,9),(-1,8)]
print(perimeter(List))
```

```python
'''Imagine two sets A and B containing numbers. Without using built-in set functionalities, write your
own function that receives two such sets and returns another set C which is a symmetric difference
of the two input sets. (A symmetric difference between A and B will return a set C which contains
only those items that appear in one of A or B. Any items that appear in both sets are not included in
C). Now compare the output of your function with the following built-in functions/operators.
✓ A.symmetric_difference(B)
✓ B.symmetric_difference(A)
✓ A ^ B
✓ B ^ A'''

def compare(A,B):
  C = set()
  for i in A:
    if i not in B:
      C.add(i)
  for j in B:
    if j not in A:
      C.add(j)
  return C

set1 = {0,1,2,3,32,1}
set2 = {3,45,56,34,2}

print(compare(set1,set2))

print(set1.symmetric_difference(set2))
print(set2.symmetric_difference(set1))
print(set1^set2)
print(set2^set1)
```

# Lab Task 3

```python
from math import pi, sin, cos

# Define step sizes
step_sizes = [0.001, 0.01, 0.1]

# Iterate through different step sizes
for h in step_sizes:
    print(f"\nResults for h = {h}\n")

    # Generate x values from -pi to pi with step size h
    x_values = [i for i in frange(-pi, pi, h)]

    # Compute derivative approximation and compare with cos(x)
    for x in x_values:
        derivative_approx = (sin(x + h) - sin(x)) / h
        actual_cos_x = cos(x)
        print(f"x: {x:.3f}, Approximated Derivative: {derivative_approx:.6f}, cos(x): {actual_cos_x:.6f}")

    print("\n" + "-" * 50)

# Helper function to generate float range
def frange(start, stop, step):
    while start < stop:
        yield start
        start += step
```
